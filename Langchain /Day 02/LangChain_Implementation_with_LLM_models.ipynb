## Langchain

- It provides tools and components for building complex AI applications, allowing developers to chain together various language model capabilities and integrate them with external data sources and APIs.

3. Ollama
- Ollama is an open-source project that allows users to run large language models locally on their own hardware.
- It provides a simple interface for running and interacting with various AI models.

image.png

```python
!pip install langchain-ollama
```

```python
from langchain_ollama.llms import OllamaLLM
model = OllamaLLM(model="gemma:2b")
our_query = "What is the currency of India?"
completion = model(our_query)
print(completion)
```
